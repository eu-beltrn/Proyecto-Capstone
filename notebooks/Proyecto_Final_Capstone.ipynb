# Introducción
En este proyecto se desarrollará una solución de ingeniería de datos para la empresa ficticia DataCommerce GT, con el objetivo de integrar información proveniente de diferentes fuentes en un único sistema confiable. A lo largo del proyecto se implementará un proceso ETL para extraer, transformar y cargar los datos en un Data Warehouse, permitiendo generar consultas, indicadores y visualizaciones que apoyen la toma de decisiones.

# Contexto de la empresa
La empresa ficticia DataCommerce GT comercializa productos tecnológicos mediante
tiendas físicas, comercio electrónico, ventas corporativas y WhatsApp. Con el
crecimiento de la empresa la información comenzó a almacenarse en diferentes
sistemas, por lo que la gerencia ya no puede obtener reportes confiables ni oportunos.
Su misión será diseñar e implementar una plataforma de datos que centralice la
información y permita generar indicadores para apoyar la toma de decisiones.

# Comprensión del negocio

La empresa DataCommerce GT comercializa productos tecnológicos mediante diferentes canales de venta. Debido al crecimiento de la empresa, la información comenzó a almacenarse en distintos sistemas independientes.
Como consecuencia existen inconsistencias entre los reportes generados por cada departamento, dificultando la toma de decisiones

# Problema empresarial

Actualmente los datos provienen de distintas fuentes, presentan errores, duplicados,
formatos diferentes y registros incompletos. Cada departamento genera reportes
distintos y los resultados no coinciden. La dirección necesita una única fuente de
información confiable.

# Objetivo General

Diseñar e implementar una solución integral de ingeniería de datos que permita integrar información de múltiples fuentes, garantizar su calidad, almacenarla en un Data Warehouse y generar información confiable para la toma de decisiones empresariales.

# Objetivos Específicos

- Integrar información proveniente de archivos CSV, Excel, JSON, SQLite y una API REST.
- Implementar un proceso ETL reproducible y automatizado.
- Aplicar reglas de limpieza y validación para garantizar la calidad de los datos.
- Diseñar un modelo relacional y un modelo dimensional.
- Construir un Data Warehouse utilizando PostgreSQL (Supabase).
- Elaborar consultas SQL para análisis empresarial.
- Generar indicadores (KPIs) mediante Python y Pandas.
- Visualizar los resultados utilizando gráficos estadísticos.
- Documentar completamente la solución desarrollada.

# Metodología de trabajo
El proyecto se desarrollará de forma incremental durante cinco días. En cada etapa se documentarán las actividades realizadas, comenzando con el análisis del negocio y la planificación de la solución, seguido de la exploración y limpieza de los datos, la construcción del Data Warehouse, el análisis mediante consultas SQL y Python, y finalmente la elaboración de la documentación y presentación del proyecto.

## Plan del proceso ETL

### Extracción (Extract)

Durante esta etapa se obtendrá la información desde las diferentes fuentes de datos proporcionadas por la empresa.

Actividades planificadas:

- Leer archivos CSV.
- Leer archivos Excel.
- Leer archivos JSON.
- Conectarse a la base de datos SQLite.
- Consumir la API REST simulada.

### Transformación (Transform)

En esta etapa se aplicarán las reglas de limpieza, estandarización y validación necesarias para garantizar la calidad de los datos antes de su almacenamiento.

Actividades planificadas:

- Eliminar registros duplicados.
- Tratar valores nulos o incompletos.
- Estandarizar formatos de texto.
- Validar tipos de datos.
- Eliminar espacios en blanco innecesarios.
- Convertir formatos de fecha.
- Normalizar nombres de columnas.
- Aplicar reglas de calidad de datos.

### Carga (Load)

Finalmente, los datos transformados serán almacenados en el Data Warehouse para su posterior análisis.

Actividades planificadas:

- Crear las tablas del Data Warehouse.
- Cargar las tablas de dimensiones.
- Cargar la tabla de hechos.
- Validar la integridad de la carga.
- Verificar la consistencia de la información almacenada.

## Alcance de la solución
La solución propuesta integrará información proveniente de diferentes fuentes de datos (CSV, Excel, JSON, SQLite y una API REST) mediante un proceso ETL que permitirá extraer, transformar y cargar los datos en un Data Warehouse. Sobre esta información consolidada se realizarán consultas SQL, análisis con Python y la generación de indicadores (KPIs) y visualizaciones para apoyar la toma de decisiones.

El alcance del proyecto comprende el diseño de la arquitectura de la solución, el modelado de datos, la implementación del proceso ETL, la carga del Data Warehouse, el análisis de la información y la documentación completa del desarrollo.
