# Introducción
En este proyecto se desarrollará una solución de ingeniería de datos para la empresa ficticia DataCommerce GT, con el objetivo de integrar información proveniente de diferentes fuentes en un único sistema confiable. # Arquitectura de la Solución – DataCommerce GT 

Para resolver la problemática de dispersión de silos de información y la inconsistencia en los reportes de **DataCommerce GT**, se diseñó una arquitectura de datos moderna basada en un **pipeline automatizado** de cuatro capas lógicas principales, complementadas con una capa final de consumo. El objetivo fundamental es garantizar la **reproducibilidad absoluta** del proceso sin intervención manual, proteger la integridad de los datos de origen y centralizar toda la información en un único repositorio de verdad.

---

# 1. Diagrama de Flujo del Dato (Pipeline End-to-End)

El flujo de datos sigue un paradigma **ETL (Extract, Transform, Load)** lineal y desacoplado, estructurado de la siguiente manera:

```text
Fuentes de Origen
        │
        ▼
Capa de Ingestión (Staging)
        │
        ▼
Capa de Procesamiento (Transformación)
        │
        ▼
Capa de Almacenamiento (Data Warehouse)
        │
        ▼
Capa de Consumo (Business Intelligence)
```

> **Nota de diseño:** Cada etapa actúa de manera independiente. Esto permite que, si un componente falla, el pipeline pueda reejecutarse desde el último punto de control sin comprometer ni corromper la base de datos analítica final.

# 2. Detalle Exhaustivo de las Capas de la Arquitectura

## Capa 1: Fuentes de Origen (Heterogeneous Data Sources)

El ecosistema actual de la empresa está compuesto por cinco fuentes de datos con diferentes estructuras, niveles de madurez digital y formatos de almacenamiento.

### Ventas
- Datos transaccionales en archivos **CSV**.
- Posibles problemas de codificación de caracteres (*encoding*).

### Productos
- Catálogos comerciales almacenados en archivos **Excel**.
- Posibles celdas combinadas y tipos de datos inconsistentes.

### Clientes
- Información almacenada en archivos **JSON**.
- Requieren procesos de **flattening** para normalizar estructuras anidadas.

### Inventario
- Datos operativos alojados en una base de datos **SQLite**.
- Acceso mediante conectores SQL nativos.

### Campañas de Marketing
- Datos obtenidos desde una **API REST simulada**.
- Requiere peticiones HTTP, validación de respuestas y manejo de posibles errores.

---

## Capa 2: Ingestión y Extracción (Staging Area & Land Zone)

### Mecanismo de viaje del dato

La extracción se realiza automáticamente utilizando Python y sus conectores especializados:

- `sqlite3`
- `requests`
- `openpyxl`

No existe manipulación manual de los archivos fuente, cumpliendo la política de mantener intactos los datos originales.

### Estrategia de Staging

Los datos son extraídos en estado **crudo (As-Is)** e importados directamente hacia **DataFrames de Pandas**.

Esta capa funciona como un **búfer de aislamiento**, permitiendo detener el proceso si ocurre alguno de los siguientes casos:

- Fallo de conexión con la API.
- Archivo bloqueado.
- Error de lectura.

De esta forma, nunca se afecta el Data Warehouse.

---

## Capa 3: Procesamiento, Limpieza y Calidad (Data Transformation Layer)

En esta capa se implementan las reglas de negocio y los controles de calidad de datos (**Data Quality Gates**).

### 1. Estandarización y Tipado

Se realizan procesos como:

- Conversión de fechas al estándar **ISO 8601 (YYYY-MM-DD)**.
- Conversión explícita de tipos de datos.
- IDs tratados como cadenas.
- Valores monetarios convertidos a tipo flotante.
- Homologación de moneda utilizando una única divisa base.

---

### 2. Limpieza de Texto y Enriquecimiento

Se aplican transformaciones orientadas a mejorar la calidad de la información:

- Eliminación de espacios dobles.
- Eliminación de caracteres especiales.
- Corrección de errores tipográficos.
- Homologación de nombres de productos.
- Homologación de nombres de clientes mediante tablas de correspondencia.

---

### 3. Validación de Integridad

Se ejecutan controles para asegurar la consistencia del modelo:

- Identificación de valores nulos.
- Imputación cuando sea justificable.
- Eliminación de registros inválidos.
- Eliminación de registros duplicados.
- Validación de llaves primarias.

---

## Capa 4: Almacenamiento Centralizado (Data Warehouse)

### Estrategia de Carga

Una vez que los datos han sido limpiados y validados, se cargan mediante procesos de **Bulk Load** hacia el Data Warehouse seleccionado:

- BigQuery
- Snowflake

---

### Evolución del Modelo de Datos

#### Esquema Relacional (Fase Inicial)

Los datos ingresan inicialmente utilizando el modelo relacional propuesto por Jonathan, manteniendo la normalización para garantizar la consistencia operativa.

---

#### Esquema Dimensional (Fase Analítica)

Posteriormente se transforman en un **Modelo Estrella (Star Schema)** orientado al análisis.

La arquitectura queda conformada por:

### Tabla de Hechos

- Fact_Ventas

### Tablas de Dimensión

- Dim_Clientes
- Dim_Productos
- Dim_Inventario
- Dim_Tiempo
- Dim_Marketing

Este diseño permite ejecutar consultas analíticas complejas con un alto rendimiento y tiempos mínimos de respuesta.

---

## Capa 5: Consumo y Analítica (Business Intelligence & Insights)

Es la capa donde la información genera valor para la organización.

Desde el Data Warehouse se publican vistas optimizadas que permiten:

- Ejecutar consultas analíticas mediante Python y Pandas.
- Calcular indicadores financieros y operativos (KPIs).
- Alimentar herramientas de visualización.
- Garantizar una única fuente oficial de información (*Single Source of Truth*).
- Eliminar discrepancias entre los reportes generados por diferentes departamentos.
